# Lets find the best 10% and worst 10% of schools and do further review

This is an imperfect way to find best schools (adding up the schools average scores), however for exploration purposes this approach will sufice because we are just looking to get a general sense of our data. 

```python
el_data['combined'] = el_data['Percentage of Grade 3 Students Achieving the Provincial Standard in Mathematics']+\
el_data['Percentage of Grade 3 Students Achieving the Provincial Standard in Reading']+\
el_data['Percentage of Grade 3 Students Achieving the Provincial Standard in Writing']+\
el_data['Percentage of Grade 6 Students Achieving the Provincial Standard in Mathematics']+\
el_data['Percentage of Grade 6 Students Achieving the Provincial Standard in Reading']+\
el_data['Percentage of Grade 6 Students Achieving the Provincial Standard in Writing']
```

```python
top = el_data['combined'].quantile(.9)
bottom = el_data['combined'].quantile(.1)
```

```python
print(top,bottom)
```

```python
top_schools = el_data[el_data['combined'] > top]
```

```python
bottom_schools = el_data[el_data['combined'] < bottom]
```

```python
sns.scatterplot(x=top_schools['Longitude'], y=top_schools['Latitude'],hue=top_schools['Percentage of Children Who Live in Low-Income Households'])
plt.title('Top School Locations')
plt.show()
```

```python
sns.scatterplot(x=bottom_schools['Longitude'], y=bottom_schools['Latitude'],hue=bottom_schools['Percentage of Children Who Live in Low-Income Households'])
plt.title('Bottom School Locations')
plt.show()
```

# where are the poorest schools?

Grab the 20% of schools with highest and lowest percentage of children from low-income households

```python
poor_thresh = el_data['Percentage of Children Who Live in Low-Income Households'].quantile(.8)
rich_thresh = el_data['Percentage of Children Who Live in Low-Income Households'].quantile(.2)
```

```python
print(poor_thresh)
print(rich_thresh)
```

```python
poor_schools = el_data[el_data['Percentage of Children Who Live in Low-Income Households'] > poor_thresh]
rich_schools = el_data[el_data['Percentage of Children Who Live in Low-Income Households'] < rich_thresh]
```

```python
poor_schools.shape
```

```python
rich_schools.shape
```

```python
sns.scatterplot(x=poor_schools['Longitude'], y=poor_schools['Latitude'])
plt.title('Location of Poorest Schools')
plt.show()
```

```python
sns.scatterplot(x=rich_schools['Longitude'], y=rich_schools['Latitude'])
plt.title('Location of Richest Schools')
plt.show()
```

## How do the poorest and richest schools compare

```python
for col in cols:
    sns.distplot(rich_schools[col],color='g', label='Rich School')
    sns.distplot(poor_schools[col], color='m', label='Poor School')
    plt.legend()
    plt.title(f'{col}')
    plt.show()
```
